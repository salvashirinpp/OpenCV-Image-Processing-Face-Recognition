# 2. Reading an Image

Reading an image means loading an image file (like JPG, PNG) into memory so it can be processed.

In OpenCV, an image is stored as a NumPy array, where:

* Each pixel contains intensity values
* A color image has 3 channels (Blue, Green, Red)

If the image path is incorrect, the image will not load.

```python
import cv2

img = cv2.imread("image.png")
```

# 5. Converting Image to Grayscale

Grayscale conversion reduces a color image into a single channel image (black and white shades).

Why it is important:

* Reduces computation
* Used in edge detection and face detection
* Removes unnecessary color information

Each pixel now represents intensity (0 = black, 255 = white).

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
```

```python
cv2.imshow("gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
_,thresh=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

cv2.imshow("gray", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 6. Edge Detection

Edge detection identifies boundaries or edges in an image.
Edges occur where there is a sudden change in intensity.

One popular method:

* Canny Edge Detection

Why important:

* Object detection
* Shape recognition
* Feature extraction

It simplifies the image by keeping only structural information.

```python
edges = cv2.Canny(img, 100, 200)
cv2.imshow("Edges", edges)
```

# 7. Cropping Image

Cropping means selecting and extracting a specific part of the image.
This is done using pixel coordinates.

Uses:

* Focus on important regions
* Remove unwanted areas
* Preprocessing for ML models

```python
crop = img[100:300, 200:400]   # [y1:y2, x1:x2]
cv2.imshow("Cropped", crop)
```

# 10. Region of Interest (ROI)

ROI means selecting a specific part of the image for processing.
Instead of working on the entire image, we focus only on a region.

Example uses:

* Face area in face detection
* Object tracking
* Medical imaging

This improves performance and reduces computation.

```python
roi = img[100:200, 100:200]
img[300:400, 300:400] = roi
```

# 12. Rotating Image

Rotation changes the orientation of an image by a specific angle.

Uses:

* Data augmentation in ML
* Correcting image alignment
* Object orientation analysis

Rotation is done using transformation matrices.

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, matrix, (w, h))
```

# 13. Drawing Shapes (Rectangle, Circle, Polylines)

**Rectangle**

Used to highlight objects (e.g., face detection box)

**Circle**

Used for marking points or features

**Polylines**

Used to draw connected lines (shapes or boundaries)

These are useful in:

* Annotation
* Visualization
* Object detection results

```python
cv2.rectangle(img, (50,50), (200,200), (0,255,0), 2)

cv2.circle(img, (150,150), 50, (255,0,0), 3)

# polylines
import numpy as np

pts = np.array([[10,10], [100,50], [200,100]], np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img, [pts], True, (0,255,255), 3)
```

# 15. Thresholding

Thresholding converts a grayscale image into a binary image (only black and white).

How it works:

If pixel value > threshold → white
Else → black

Uses:

Image segmentation
Object separation
OCR (text detection)

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
cv2.imshow("Threshold", thresh)
```

# 16. Playing Video

A video is a sequence of frames (images).
OpenCV reads and displays video frame by frame.

Uses:

* Video analysis
* Surveillance systems
* Motion detection

Each frame is processed like an image.

```python
cap = cv2.VideoCapture("video.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    cv2.imshow("Video", frame)
    
    if cv2.waitKey(25) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# 17. Capturing Video from Camera

This involves accessing the webcam to capture live video.

Uses:

* Real-time face detection
* Gesture recognition
* Security systems

The camera continuously provides frames for processing.

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    
    cv2.imshow("Camera", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# 18. Saving Video

Saving video means recording frames and storing them as a video file.

Important components:

* Codec → defines format (e.g., AVI, MP4)
* Frame rate → frames per second
* Resolution → size of video

Used in:

* CCTV recording
* Screen capture
* Data collection

```python
cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480))

while True:
    ret, frame = cap.read()
    
    out.write(frame)
    cv2.imshow("Recording", frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

recording = False

while True:
    ret, frame = cap.read()
    if not ret:
        break

    cv2.imshow("Webcam", frame)

    key = cv2.waitKey(1)

    if key == ord('s'):
        recording = True
        print("Recording started...")

    elif key == ord('q'):
        print("Recording stopped.")
        break

    if recording:
        out.write(frame)

cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
face_cascade=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    cv2.putText(frame, f'Faces: {len(faces)}', (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow("Face Detection", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
